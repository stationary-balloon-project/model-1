As high as the balloon gets, lower is the air pressure, density and temperature. All these variables are important to calculate the balloon properties.

The equation for computing pressure as function of height above the sea level and below 86km is:

`P = Pb * ((Tb + (h - hb) * Lb) / Tb) ^ ((-g * M) / (R * Lb))`, where:
- `P` is the calculated pressure (Pa)
- `Pb` is the reference pressure (Pa)
- `Tb` is the reference temperature (K)
- `Lb` is the temperature lapse rate (K/m)
- `h` is the height of calculation (m)
- `hb` is the height of reference level b (m)
- `R` is the universal gas constant (J/(mol * K))
- `g` is the gravitational acceleration (m/s^2)
- `M` is the molar mass of Earth's air (kg/mol)

The equation for computing density as function of height above the sea level and below 86km is:

`p = pb * (Tb / (Tb + (h - hb) * Lb)) ^ (1 + ((g * M) / R * Lb))`, where:
- `p` is the mass density (kg/m^3)
- `pb` is the reference mass density (kg/m^3)
- `Tb` is the standard temperature (K)
- `Lb` is the temperature lapse rate (K/m)
- `h` is the height of calculation (m)
- `hb` is the height of reference level b (m)
- `R` is the universal gas constant (J/(mol * K))
- `g` is the gravitational acceleration (m/s^2)
- `M` is the molar mass of Earth's air (kg/mol)
