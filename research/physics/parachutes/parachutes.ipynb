On the begining of the falling moviment, the body velocity is 0. The only force acting on this object is the gravitatioal force. So, as the gravitational force is the only force acting on the object, its also the resultant force.

`V = 0 ; P = Fres`

As the body fdlls, a opposing, force proportional to the objects falling velocity, takes place. It's called air resistence, or drag, force.

![](forces.png "Forces")

In the above image, as `V` increases, `P > Fd`, so there's a resultant force perpendicular to `P`, that is less than `Fres` when `V = 0`.

![](forces-2.png "Forces")

As `V` increases, `Fd` increases too, to the point where `P = Fd`. At this moment, `Fres = 0`, so there's no more acceleration to the body, then keeping it's `V` constant according to the first Newton's law, which states that an object at rest or in uniform motion in a straight line tends to remain in that state if the resultant force on it is zero.
This moment, where `Fd = 0` and `V` becomes constant, this velocity is called a terminal velocity.

When the parachute is opened, `Fd` increases, resulting in a `Fres > 0` perpendicular to `Fd`, decreasing `V`, that is perpendicular to `P`.

![](forces-3.png "Forces")

As `V` decreases, `Fd` decreases proportionaly, until the moment where `Fd = P` again, but this time `V` is less than before.

![](forces-4.png "Forces")

To calculate `Fd`, we use `Fd = 0.5 * p * c * A * V^2`, where:
- `p` is the density of the air
- `c`, the drag coefficient
- `A`, the parachute area
- `V`, the falling velocity

Also, we have `Fres = P + Fd`.

When the body reaches it's terminal velocity, `P = Fd`, and `P = m * g`, being `m` the mass of the body and `g` the gravity's acceleration, so `m * g = 0.5 * p * c * A * V^2`.
We can derivate this formula to calculate the terminal velocity as follows:
- `m * g = (1/2) * p * c * A * V^2`
- `2m * g = p * c * A * V^2`
- `2m * g / (p * c * A) = V^2`

Now derivating the parachute area:
- `m * g = (1/2) * p * c * A * V^2`
- `2m * g = p * c * A * V^2`
- `A = 2m * g / (p * c * V^2)`
